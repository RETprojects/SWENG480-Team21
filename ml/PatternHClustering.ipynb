```python
# Step 1: load the data
import pandas as pd
df = pd.read_csv("scraped_pattern_data.csv")
```

```python
# Step 2: Explore the data
df.info()
```

```python
# Step 3: Data preprocessing
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = df['overview'].values.astype("U")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
k = df.count()[0]
model = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=1)
model.fit(features)
```

```python
df['cluster'] = model.labels_
```

```python
df.head()
```

```python
from sklearn import cluster
# output the results to a text file.
clusters = df.groupby('cluster')

for cluster in clusters.groups:
  f = open('cluster'+str(cluster)+'.csv', 'w') #create csv file
  data = clusters.get_group(cluster)[['name', 'overview']] # get title and overview colmns
  f.write(data.to_csv(index_label='id')) #set index to id
  f.close()
```

```python
print("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

for i in range(k):
  print("Cluster %d:" % i)
  for j in order_centroids[i, :10]: #print out 10 features terms of each cluster
    print('%s'% terms[j])
  print('------------------')
```
